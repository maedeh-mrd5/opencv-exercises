```python
# Import libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the original image
img_original = cv.imread('images/result_swap.jpeg')

# Compute the center coordinates of the image
x_center, y_center = img_original.shape[0] // 2, img_original.shape[1] // 2

# Extract the top-left quadrant of the image
img_top_left = img_original[:x_center,:y_center,:]

# Extract the top-right quadrant of the image
img_top_right = img_original[:x_center,y_center:,:]

# Extract the bottom-left quadrant of the image
img_bottom_left = img_original[x_center:,:y_center,:]

# Extract the bottom-right quadrant of the image
img_bottom_right = img_original[x_center:,y_center:,:]

# Create a new empty image with the same shape and data type as the original
img = np.zeros(img_original.shape, np.uint8)

# Place the bottom-right quadrant into the top-left position
img[:x_center,:y_center,:] = img_bottom_right

# Place the top-right quadrant into the bottom-left position
img[x_center:,:y_center,:] = img_top_right

# Place the top-left quadrant into the bottom-right position
img[x_center:,y_center:,:] = img_top_left

# Place the bottom-left quadrant into the top-right position
img[:x_center,y_center:,:] = img_bottom_left

# Display images
plt.figure(figsize=[18, 12])
plt.subplot(231); plt.imshow(img_original[...,::-1]); plt.title('Original Image')
plt.subplot(232); plt.imshow(img_top_left[...,::-1])
plt.subplot(233); plt.imshow(img_top_right[...,::-1])
plt.subplot(234); plt.imshow(img_bottom_left[...,::-1])
plt.subplot(235); plt.imshow(img_bottom_right[...,::-1])
plt.subplot(236); plt.imshow(img[...,::-1])
```
